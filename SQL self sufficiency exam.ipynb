```python
# 3) Write a query that returns summary statistics for avg_math_4_score by state. 
#    Make sure to sort the results alphabetically by state name.

sql = '''
SELECT  MIN(avg_math_4_score) AS min_math_4_score, state
FROM naep
WHERE avg_math_4_score IS NOT NULL 
GROUP BY state
ORDER BY state'''

max_avg_4_math_score = engine.execute(sql)
engine.dispose()

for rows in max_avg_4_math_score:
    print(rows)
```

```python
# 4) Write a query that alters the previous query so that it returns only the summary statistics for avg_math_4_score 
#    by state with differences in max and min values that are greater than 30

sql ='''
SELECT  ROUND(MAX(avg_math_4_score) - MIN(avg_math_4_score),2) AS "diff_greater_30" , state
FROM naep
GROUP BY state 
HAVING  MAX(avg_math_4_score) - MIN(avg_math_4_score) > 30
ORDER BY state
'''

greater_than_30 = engine.execute(sql)
engine.dispose()

for rows in greater_than_30:
    print(rows)
```

```python
# 7) Write a query that returns a field called below_average_states_y2000 
#    that lists all states with an avg_math_4_score less than the average 
#    over all states in the year 2000.

sql = '''
SELECT avg_math_4_score AS below_average_states_y2000, state, year
FROM naep
WHERE avg_math_4_score <

 
  (SELECT ROUND(AVG(avg_math_4_score),2) 
   FROM naep
   WHERE year = 2000 and avg_math_4_score IS NOT NULL
   GROUP BY year )
ORDER BY state,year
'''

below_average_states_y2000 = engine.execute(sql)
engine.dispose()

for rows in below_average_states_y2000:
    print(rows)
```

```python
# 9) Write a query that returns for the year 2000 the state, avg_math_4_score, 
#    and total_expenditure from the naep table left outer joined with the finance 
#    table, using id as the key and ordered by total_expenditure greatest to least. 
#    Be sure to round avg_math_4_score to the nearest 2 decimal places, and then 
#    filter out NULL avg_math_4_scores in order to see any correlation more clearly.

sql = '''
SELECT finance.state, ROUND((avg_math_4_score),2) AS avg_math_4_score ,total_expenditure
FROM naep LEFT OUTER JOIN finance
ON naep.id = finance.id 
WHERE naep.year = 2000 AND avg_math_4_score IS NOT NULL
GROUP BY finance.state, avg_math_4_score, total_expenditure
ORDER BY avg_math_4_score DESC
'''

expenditure_correlation = engine.execute(sql)
engine.dispose()

for rows in expenditure_correlation:
    print(rows)
```
